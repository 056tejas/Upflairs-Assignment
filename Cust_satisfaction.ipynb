```python
import pandas as pd
```

```python
df=pd.read_csv("cust_satisfaction.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median())
```

```python
gender_map = {'Male': 0, 'Female': 1}
customer_type_map = {'Loyal Customer': 0, 'disloyal Customer': 1}
travel_type_map = {'Business travel': 0, 'Personal Travel': 1}
class_map = {'Eco': 0, 'Eco Plus': 1, 'Business': 2}
satisfaction_map = {'neutral or dissatisfied': 0, 'satisfied': 1}
```

```python
df['Gender'] = df['Gender'].map(gender_map)
df['Customer Type'] = df['Customer Type'].map(customer_type_map)
df['Type of Travel'] = df['Type of Travel'].map(travel_type_map)
df['Class'] = df['Class'].map(class_map)
df['satisfaction'] = df['satisfaction'].map(satisfaction_map)
```

```python
df['Total Delay'] = df['Departure Delay in Minutes'] + df['Arrival Delay in Minutes']
```
