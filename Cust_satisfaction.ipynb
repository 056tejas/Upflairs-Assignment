```python
import pandas as pd
```

```python
df=pd.read_csv("cust_satisfaction.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median())
```

```python
gender_map = {'Male': 0, 'Female': 1}
customer_type_map = {'Loyal Customer': 0, 'disloyal Customer': 1}
travel_type_map = {'Business travel': 0, 'Personal Travel': 1}
class_map = {'Eco': 0, 'Eco Plus': 1, 'Business': 2}
satisfaction_map = {'neutral or dissatisfied': 0, 'satisfied': 1}
```

```python
df['Gender'] = df['Gender'].map(gender_map)
df['Customer Type'] = df['Customer Type'].map(customer_type_map)
df['Type of Travel'] = df['Type of Travel'].map(travel_type_map)
df['Class'] = df['Class'].map(class_map)
df['satisfaction'] = df['satisfaction'].map(satisfaction_map)
```

```python
df['Total Delay'] = df['Departure Delay in Minutes'] + df['Arrival Delay in Minutes']
```

```python
df.head()
```

```python
df.groupby('Class')['satisfaction'].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression

from sklearn.metrics import accuracy_score, mean_squared_error
```

```python
X = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
lin_pred=lin_reg.predict(X_test)
```

```python
lin_pred
```

```python
print("training score:- ",lin_reg.score(X_train, y_train))
```

```python
print("testing score:- ",lin_reg.score(X_test, y_test))
```

```python
lin_mse = mean_squared_error(y_test, lin_pred)
```

```python
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)
```

```python
log_pred=log_reg.predict(X_test)
```

```python
log_pred
```

```python
print("training score:- ",log_reg.score(X_train, y_train))
```

```python
print("testing score:- ",log_reg.score(X_test, y_test))
```

```python
log_acc = accuracy_score(y_test, log_pred)
```

```python
print("Linear Regression MSE:", lin_mse)
print("Logistic Regression Accuracy:", log_acc)
```
